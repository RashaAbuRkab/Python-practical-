- <font color='blue'>Exercise# 1:</font> Write a program that reads in the name and salary of an employee. Here the salary will denote an hourly wage, such as $9.25. Then ask how many hours the employee worked in the past week. Be sure to accept fractional hours. Compute the pay. Any overtime work (over 40 hours per week) is paid at 150 percent of the regular wage. Print a paycheck for the employee.  
  
|<strong>Sample Run:</strong>|<strong>Sample Run:</strong>|
| :------------------------- | :------------------------- |
| Enter your name: Ahmad     | Enter your name: Ali       |
| Enter your salary: 23.3    | Enter your salary: 45.7    |
| Enter the number of worked hours: 43.5|Enter the number of worked hours: 20.45|
| Mr. Ahmad, your payment is 1054.33|Mr. Ali, your payment is 934.57|

````python
# Exercise # 1 - Source Code

Name=input("Enter your name: ")
Salary=float(input("Enter your salary: "))
Hours=float(input("Enter the number of worked hours: "))

if Hours > 40:
    payment=(Salary*40)+((Hours-40)*(Salary*1.5))
else:
    payment=Salary*Hours

print("Mr. %s, your payment is %.2f"%(Name,payment))
````

- <font color='blue'>Exercise# 3:</font> 
Write a program that reads a temperature and its unit (C or F) from the user, then it converts it to the other unit as shown in the sample runs.
The following is the formula for converting the temperature from Celsius to Fahrenheit.<br>
C = (F -32) * 9/5 <br>
$$Fahrenheit=32+\frac{9}{5}*Celsius$$
Note: the input can be lower case or upper case

````python
# Exercise # 3 - Source Code

temperature=float(input("Enter current temperature:"))
Unit=input("Enter it's unit C or F:")
if Unit=="C" or Unit=="c":
    temperatureC=temperature
    temperatureF=(temperatureC * (9/5)) + 32
    print("%.2f C = %.2f F"%(temperatureC,temperatureF))
elif Unit=="F" or Unit=="f":
    temperatureF=temperature
    temperatureC = (temperatureF - 32) * (5/9)
    print("%.2f F = %.2f C"%(temperatureF,temperatureC))
else:
    print("Wrong unit")
````

- <font color='blue'>Exercise# 4:</font> Write a program that displays the following menu:

*Note:  Assume that the values read by the program for options 1, and 2 are in centimeters. For these options, your program must display appropriate units in the output as shown below.*  
  
  


````python
# Exercise # 4 - Source Code

import math
print("1: Rectangle")
print("2: Circle")
choice=int(input("Enter your choice: "))
if (choice==1):
    width=float(input("Enter width: "))
    length=float(input("Enter length: "))
    area=width*length
    print("Rectangle area= %.2f cm^2"% area)
elif (choice==2):
    radius=float(input("Enter radius: "))
    area=math.pi*radius*radius
    print("Circle area= %.2f cm^2"% area)
else:
    print("Wrong choice")
````
