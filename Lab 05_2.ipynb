## Exercise # 2:
Write a program that asks the user for a collection of $N$ numbers. Then, it will find the smallest value and its frequency.
- Get the value of $N$ from the user.
- If $N \le 0$, display an appropriate error message and terminate the program; otherwise
- Read the values as entered from the user. (If $N =5$, then there are 5 values the user is going to enter).
- Find the smallest, its frequency (how many times it is repeated). 

- **Hint**: To find the smallest value, read the first value **before the loop** then assume that this first value entered by the user is the smallest value. **Within the loop** compare the assumed smallest value with other values **read inside the loop** and change the smallest value accordingly.  

### Two sample runs



```python
#%%writefile lab07Ex2.py
## Uncomment the above line after you finish your code and want to save it in a file.

# YOUR CODE HERE

number = int(input("Enter how many values to consider: "))

if number <= 0 :
    print("Wrong input. Input must be > 0 ")
else:
    minValue = int(input("Enter value 1: "))    
    frequency = 1
       
    for x in range(2,number+1):
        newvalue = int(input("Enter value %d: "%x))
           
        if newvalue < minValue :
            minValue = newvalue
            frequency = 1
               
        elif newvalue == minValue :
            frequency = frequency + 1
   
    print("*"*15)
    print("Minimum value = ", minValue)
    print("It's frequency is", frequency)
```

## Exercise # 3:
Write a program that reads a string then generates and prints a new string that has same characters of the original string but with alternating capital and small letters. For example, check the sample RUN below.

### Sample run



```python
#%%writefile lab07Ex3.py
## Uncomment the above line after you finish your code and want to save it in a file.

# YOUR CODE HERE

str = input("Enter a statment: ")
tempWord = str.lower()
outstr=""
for ind in range(0,len(tempWord)):
   
    if ind%2==0:
        outstr=outstr+tempWord[ind].upper()
    else:
        outstr=outstr+tempWord[ind]

print("you mean:")
print(outstr)
```

## Exercise # 4:
Write a program that reads a word and prints the number of vowels in the word. For this exercise, assume that **a e i o u y** are vowels. For example, if the user provides the input "Harry", the program prints 2 vowels.  
**Sample Runs**

```python
#%%writefile lab07Ex4_YourID.py
## Uncomment the above line after you finish your code and want to save it in a file.

# YOUR CODE HERE

word = input("Enter a word: ")
tempWord = word.lower()
vowels = 0
for char in tempWord:
    if "a"==char or "o"==char or "u"==char or "i"==char or "e"==char or "y"==char:
        vowels = vowels + 1
        
print("There are", vowels, "vowels in", word)
```
